```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import seaborn as sns
import datetime as dt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings('ignore')


%matplotlib inline
```

```python
#Clean CSV File
movie_data_to_load = "Output/new_movie_df_cleaned.csv"
new_movies_df = pd.read_csv(movie_data_to_load,index_col = 0)
```

Question 1:
Do profitable movies tend to have higher or lower average review scores (on a scale of 1 to 10) than unprofitable movies?

```python
#DataFrames for profitable and not profitable movies
profitable_movies = new_movies_df[new_movies_df["Profitability"]=="Profitable"]
not_profitable_movies = new_movies_df[new_movies_df["Profitability"]=="Not Profitable"]
```

```python
sns.distplot(not_profitable_movies["Average Vote"], color = "orange", label = "Not Profitable", bins = 10)
sns.distplot(profitable_movies["Average Vote"], color = "blue",label = "Profitable", bins = 10)

plt.title("Film Ratings Distribution")
plt.xlabel("Rating (1 to 10)")
plt.legend(fontsize="small", loc = "upper left")
sns.despine()
plt.tight_layout()

plt.savefig("Figures/film_ratings_distribution.png")
plt.show()
```

```python

ratings_boxplot=sns.boxplot(x="Profitability",y="Average Vote",data=new_movies_df, palette="coolwarm", order=["Not Profitable","Profitable"], boxprops=dict(alpha=0.6))


ratings_boxplot.artists[0].set_facecolor("orange")
ratings_boxplot.artists[1].set_facecolor("blue")


plt.axes().yaxis.grid(True,color="grey",linestyle="--",alpha=0.5)
sns.despine()
plt.xlabel("")
plt.ylabel("Rating (1 to 10)")
plt.title("Average Film Ratings")
plt.tight_layout()


plt.savefig("Figures/film_ratings_boxplot.png")
plt.show()
```

```python
# Check variances of average review scores of both profitable and unprofitable films
{"Unprofitable Film Rating Standard Deviation":(np.std(not_profitable_movies["Average Vote"])),"Profitable Film Rating Standard Deviation":(np.std(profitable_movies["Average Vote"]))}
```

Question 2:
Do profitable movies tend to have higher or lower budgets than unprofitable movies?

```python
#Budget Distribution


sns.distplot(not_profitable_movies["Budget"], color = "orange", label = "Not Profitable", bins = 10)
sns.distplot(profitable_movies["Budget"], color = "blue",label = "Profitable", bins = 10)


plt.title("Film Budget Distribution")
plt.xlabel("Budget (in hundred million USD)")
plt.legend(fontsize="small", loc = "upper right")
sns.despine()
plt.tight_layout()

plt.savefig("Figures/film_budget_distribution.png")
plt.show()
```

```python
budget_boxplot= sns.boxplot(x="Profitability",y="Budget",data=new_movies_df,palette="coolwarm",order=["Not Profitable","Profitable"], boxprops=dict(alpha=0.6))


budget_boxplot.artists[0].set_facecolor("orange")
budget_boxplot.artists[1].set_facecolor("blue")


plt.axes().yaxis.grid(True,color="grey",linestyle="--",alpha=0.5)
sns.despine()
plt.xlabel("")
plt.ylabel("log (Budget (USD))")
plt.title("Film Budgets")
plt.tight_layout()


plt.savefig("Figures/log_film_budget_boxplot.png")
plt.show()
```
